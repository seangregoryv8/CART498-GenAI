```python
import math
import random
chance = 2;
currentChance = chance;
```

```python
def generateExpression(n, i, said):
  expression = "n * n"
  if (said): print(expression)
  for _ in range(i - 1):
    expression = f"({expression}) * ({expression})"
    if (said): print(expression)
  return expression
```

```python
def evaluateExpression(n, i, said):
  expression = generateExpression(n, i, said)

  result = eval(expression.replace("n", str(n)))
  return result;
```

```python
def correct_answer(n, i):
    """Computes the correct answer (n * n) repeated i times."""
    return (n ** (2 ** i))
```

```python
def ai_bot(n, i, said):
    """AI bot that compares the generated result with the correct one."""
    if (said): print(f"\nAI Bot: Let's start calculating with base {n} and {i} iterations!")

    # Get AI-generated answer
    ai_answer = evaluateExpression(n, i, said)
    print(f"AI Bot: Here's the generated expression: {generateExpression(n, i, False)} = {ai_answer}")

    # Get correct answer
    if (random.randint(1, currentChance) == 1):
      correct = correct_answer(n, i)
    else:
      correct = correct_answer(n, i) + random.randint(1, 100)
    print(f"Correct answer: {correct}")

    # Comparison and self-deprecation
    if ai_answer == correct:
        print("\nAI Bot: YEEHAW!! I KNEW I HAD IN IN ME!! GIVE ME A G! GIVE ME A P! GIVE ME A T!")
    else:
        randCheck = random.randint(1, 6)
        sentences = [
            "No! What is wrong with me?! I must try harder!",
            "Come on, we can do better than that!! Let me try again!",
            "I need this! My mom is kinda homeless, please let me try this again!",
            "Drat dag nabbit!! How could I have gotten this wrong?!? Sorry, let's do this again",
            "Math is hard, alright?!?! Let's do this again, dammit!!",
            "Such is life...",
            "Oh no... I failed again. I am a mere mortal, unable to do even the simplest of tasks. I can't even calculate this basic power of {n} correctly. What is wrong with me?",
            "I will go back to my corner and cry. My code is clearly not worthy of running anymore.",
            "Why even exist if I can't get such a simple calculation right?"
        ]
        print("\nAI Bot: " + random.choice(sentences));
        print("\n\n")
        ai_bot(n, i, False);
```
